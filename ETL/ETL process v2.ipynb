```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("ETL-process").config("spark.driver.extraClassPath","/content/postgresql-42.2.16.jar").getOrCreate()
```

```python
from pyspark import SparkFiles
url = "https://flightsproject.s3.us-east-2.amazonaws.com/flights_2021.csv"
spark.sparkContext.addFile(url)
df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get(""), sep="\t", header=True, inferSchema=True)
df.show()
```
