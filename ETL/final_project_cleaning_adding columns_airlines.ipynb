### ETL - Loading Data 

```python
import pandas as pd
```

```python
#exporting data from CSV
flights_airlines_df=pd.read_csv('https://flightsproject.s3.us-east-2.amazonaws.com/flights_city_names.csv')
flights_airlines_df
```

```python
#exporting data from CSV
airlines_cat_df=pd.read_csv('https://flightsproject.s3.us-east-2.amazonaws.com/Airlines.csv')
airlines_cat_df
```

```python
#in dictionary
abbr_companies = airlines_cat_df.set_index('Code')['Description'].to_dict()
```

```python
abbr_companies
```

```python
# I extract a subset of columns and redefine the airlines labeling 
flights_name_df = airlines_cat_df.loc[:, ['Description']]
flights_name_df['Description'] = flights_name_df['Description'].replace(abbr_companies)
```

```python
flights_name_df
```

```python
#rename code
#abbr_companies = airlines_cat_df.set_index('Code')['Description'].to_dict()
airlines_cat_df.rename(columns = {'Code':'IATA_Code_Marketing_Airline'}, inplace = True)
airlines_cat_df
```

```python
airlines_cat_df.rename(columns = {'Description':'Airline'}, inplace = True)
airlines_cat_df
```

```python
#merge
flights_merged_df = flights_airlines_df.merge(airlines_cat_df,on=["IATA_Code_Marketing_Airline"])
flights_merged_df
```

```python
flights_data_df = flights_merged_df.drop(['Unnamed: 0', 'IATA_Code_Marketing_Airline', 'Operating_Airline '], axis = 1)
flights_data_df
```

```python
cols = list(flights_data_df.columns.values)
cols
```

```python
flights_data_df = flights_data_df[['Year', 'Month', 'DayofMonth', 'DayOfWeek', 'Airline', 'Origin', 'OriginCityName', 'Dest', 'DestCityName', 'CRSDepTime', 'DepTime', 'DepDelayMinutes', 'CRSArrTime' ,'ArrTime', 'Cancelled', 'CancellationCode', 'CRSElapsedTime', 'ActualElapsedTime', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay', 'FirstDepTime']]
flights_data_df
```

```python
flights_data_df.to_csv('flights_data_df.csv')
```
