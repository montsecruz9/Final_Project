```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
```

```python
#  Import and read the flights_data.csv.
flights_df = pd.read_csv("flights_data_df.csv")
flights_df.head()
```

```python
flights_df["Is_Delayed"] = None
flights_df.loc[flights_df["DepDelayMinutes"] == 0, "Is_Delayed"] = "1"
flights_df.loc[flights_df["DepDelayMinutes"] > 0, "Is_Delayed"] = "0"
```

```python
flights_df.head()
```

```python
# Dropping columns
flights_df.drop(['Origin', 'Dest', 'Unnamed: 0', 'Cancelled', 'CarrierDelay', 'WeatherDelay', "NASDelay", 'SecurityDelay', 'LateAircraftDelay', 'FirstDepTime', 'CancellationCode'], axis=1)
```

```python
flights_df.head()
```

```python
pd.get_dummies(flights_df, columns=['Operating_Airline '])
```

```python
# Generate our categorical variable lists
flights_cat = flights_df.dtypes[flights_df.dtypes == "object"].index.tolist()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(flights_df[flights_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(flights_cat)
encode_df.head()
```
